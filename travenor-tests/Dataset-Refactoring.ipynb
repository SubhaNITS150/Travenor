```python
import pandas as pd
from geopy.geocoders import Nominatim
from tqdm import tqdm
import time
import os
```

```python
path = r"C:/Users/subha/Projects/travenor-tests/dataset/destinations.csv"
# df = pd.read_csv(path)
print(os.path.exists(path))
```

```python
df = pd.read_csv(path)
df
```

```python
geolocator = Nominatim(user_agent="travenor")
```

```python
latitudes = []
longitudes = []
```

```python
for _, row in tqdm(df.iterrows(), total=len(df)):
    query = f"{row['Name']}, {row['City']}, {row['State']}, India"

    try:
        location = geolocator.geocode(query)

        if location:
            latitudes.append(location.latitude)
            longitudes.append(location.longitude)
        else:
            latitudes.append(None)
            longitudes.append(None)

        time.sleep(1)  

    except Exception as e:
        print("Error:", e)
        latitudes.append(None)
        longitudes.append(None)
```

```python
df["latitude"] = latitudes
df["longitude"] = longitudes
```

```python
df.to_csv("destinations_with_coordinates.csv", index=False)

print("Done")
```
