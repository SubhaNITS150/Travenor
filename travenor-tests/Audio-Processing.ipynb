### Applying Whisper 

```python
cleaned_audio = r"C:/Users/subha/Projects/travenor-tests/cleaned/converted10_DeepFilterNet3.wav"
```

```python
import torch
import whisperx
import os

# --- Define your audio path here ---
cleaned_audio = r"C:\Users\subha\Projects\Travenor\travenor-tests\cleaned\converted10_DeepFilterNet3.wav"

# Validate the file exists before doing anything
if not os.path.exists(cleaned_audio):
    raise FileNotFoundError(
        f"Audio file not found: {cleaned_audio}\n"
        f"Please check:\n"
        f"  1. The file path is correct\n"
        f"  2. The file was actually saved/generated by the previous step\n"
        f"  3. You're using the right filename and extension"
    )

print(f"Audio file found: {cleaned_audio}")

device = "cpu"

model = whisperx.load_model(
    "medium",
    device=device,
    compute_type="int8"
)

print("WhisperX model loaded!")
result = model.transcribe(cleaned_audio)
print("Transcription completed!")
print(result)
```

```python
transcript = "" 
for segment in result["segments"]:
    print(segment["text"])
    transcript += segment["text"]
```

### Diarization model

```python
diarize_model = whisperx.DiarizationPipeline(
    device=device
)

print("Diarization model loaded!")
```

```python
diarize_segments = diarize_model(cleaned_audio)

print("Diarization completed!")
```

```python
result = whisperx.assign_word_speakers(
    diarize_segments,
    result
)

print("Speakers assigned!")
```

#### Speaker wise script assignment

```python
for segment in result["segments"]:
    
    speaker = segment.get("speaker", "UNKNOWN")
    text = segment["text"]

    print(f"{speaker}: {text}")
```

```python
from gliner import GLiNER
```

```python
model = GLiNER.from_pretrained(
    "urchade/gliner_medium-v2.1"
)

labels = [
    # Geography
    "location",
    "place",
    "city",
    "state",
    "country",
    "village",
    "tourist destination",
    "mountain",
    "trek",
    "region",
    
    # Natural Tourist Spots
    "beach", "mountain", "national park", "lake", "island",
    
    # Built Tourist Spots
    "temple", "museum", "monument", "historical site",
    
    # Catch-all for other specific attractions (e.g., "Eiffel Tower")
    "tourist attraction"
]
entities = model.predict_entities(
    transcript,
    labels
)

for e in entities:
    print(e)
```

```python
destinations = [entity["text"] for entity in entities]

unique_destinations = list(dict.fromkeys(destinations))

print(unique_destinations)
```

## Generating Cluster

```python
%pip install networkx python-louvain matplotlib --quiet
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time
```

```python
geolocator = Nominatim(user_agent="travel_recommender")
locations = []
```

```python
for place in destinations:

    try:
        location = geolocator.geocode(place)

        if location:
            locations.append({
                "name": place,
                "latitude": location.latitude,
                "longitude": location.longitude
            })

        time.sleep(1)

    except Exception as e:
        print(place, e)
```

```python
extracted_locations = pd.DataFrame(locations)

extracted_locations
```

### Visualization on map

```python
import folium
```

```python
center_lat = extracted_locations.latitude.mean()
center_lon = extracted_locations.longitude.mean()

m = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=7
)

for _, row in extracted_locations.iterrows():

    folium.Marker(
        [row.latitude, row.longitude],
        popup=row.name
    ).add_to(m)

m
```

```python
import numpy as np
from sklearn.cluster import DBSCAN

coords = np.array(
    list(
        zip(
            extracted_locations.latitude,
            extracted_locations.longitude
        )
    )
)
```

```python
kms_per_radian = 6371.0088

epsilon = 100 / kms_per_radian

db = DBSCAN(
    eps=epsilon,
    min_samples=1,
    metric="haversine"
)

labels = db.fit_predict(
    np.radians(coords)
)
```

```python
extracted_locations["cluster"] = labels

extracted_locations
```

```python
tour_groups = {}

for cluster_id in extracted_locations["cluster"].unique():

    places = extracted_locations[
        extracted_locations["cluster"] == cluster_id
    ]["name"].tolist()

    tour_groups[cluster_id] = places

tour_groups
```
